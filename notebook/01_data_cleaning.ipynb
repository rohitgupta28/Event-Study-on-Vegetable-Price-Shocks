```python
import pandas as pd
from pathlib import Path

RAW_PATH = Path("../data/halflife_q.xlsx")
CLEAN_PATH = Path("../data/clean_panel.csv")
```

```python
df = pd.read_excel(RAW_PATH, sheet_name="Sheet1")
df.head()
```

```python
df.info()
df.describe(include='all')
for c in df.columns:
    print(c, df[c].nunique())
```

```python
panel = df[['State', 'Year', 'Month', 'relc']].copy()
month_map = {
    "january":1,"february":2,"march":3,"april":4,"may":5,"june":6,
    "july":7,"august":8,"september":9,"october":10,"november":11,"december":12
}

panel['Month_num'] = (
    panel['Month'].str.lower().map(month_map)
)


panel['Date'] = pd.to_datetime(
    dict(year=panel['Year'], month=panel['Month_num'], day=1)
)

panel = panel.sort_values(['State','Date']).reset_index(drop=True)
panel.to_csv("../data/clean_panel.csv", index=False)
panel.head()
```
